```python
class human():
    def __init__(self, age, weight):
        self.eye = 2
        self.nose = 1
        self.ear = 2
        self.mouth = 1
        self.age = age
        self.weight = weight

    def say(self, msg):
        print(f'{self.name} say;{msg}')
    
    def play(self,thing):
        print(thing)

human.hand = 2
human.leg = 2

oxxo = human(18, 45)
gkpen = human(20, 50)


print(f'oxxo, age; {oxxo.age}, weight; {oxxo.weight}')
print(f'gkpen, age: {gkpen.age}, weight: {gkpen.weight}')
```

```python
class father():
    def __init__(self):
        self.eye = 2
        self.ear = 2
        self.nose = 1
        self.mouth = 1
        
class mother():
    def language(self):
        print('English')
    def skill(self):
        print('painting')

class son(father, mother):
    def play(self):
        print('balls')

oxxo = son()
print(oxxo.eye)    
oxxo.play()
oxxo.skill()
oxxo.language()  
```
